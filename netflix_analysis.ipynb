# Netflix Data Analysis.

This project analyzes Netflix content using Python

Tools used:
-Pandas
-Matplotlib
-Seaborn

Goal:
To understand Netflix content trends like ratings, countries, and show types.

## Import Libraries

```python
## Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
## Load Dataset
df = pd.read_csv("Data/netflix_titles 2.csv")
df.head()
```

## Movies VS TV Shows on Netflix

```python
sns.countplot(x='type' , data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
          
```

## Top 10 Countries Producing Netflix Content

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar')
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Shows")
plt.show()
```

## Top Release Years on Netflix

```python
df['release_year'].value_counts().head(10).plot(kind='bar')
plt.title("Top Release Years on Netflix")
plt.xlabel("Year")
plt.ylabel("Number of Shows")
plt.show()
```

## Netflix Content Added over the Years

```python
df['date_added'] = pd.to_datetime(df['date_added'],errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title("Netflix Content Added over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Shows")
plt.show()
```

## Top 10 Genres on Netflix

```python
genres = df['listed_in'].str.split(',',expand=True).stack()
top_genres = genres.value_counts().head(10)
top_genres.plot(kind='bar')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genres")
plt.ylabel("Number of titles")
plt.show()
```

## Most Common Ratings on Netflix

```python
df['rating'].value_counts().head(10).plot(kind='bar')
plt.title("Most Common Rating on Netflix")
plt.xlabel("Rating")
plt.ylabel("Number of Shows")
plt.show()
```
